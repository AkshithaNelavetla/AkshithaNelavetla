```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pdsdf = pd.read_csv( r"/content/StudentsPerformance.csv" )
```

```python
pdsdf
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='test preparation course', data=pdsdf, palette='rocket', order=pdsdf['test preparation course'].value_counts().index) # Changed df to pdsdf to match the DataFrame variable name
plt.xticks(rotation=45, ha='right')
plt.xlabel('test preparation course')
plt.ylabel('Count')
plt.title('Bar Chart of test preparation course Types')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(pdsdf['reading score'], bins=20, kde=True, color='tan') # Changed df to pdsdf
plt.xlabel('reading Score')
plt.ylabel('Frequency')
plt.title('Histogram of reading Scores')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='math score', data=pdsdf, palette='viridis') # Changed df to pdsdf to match the DataFrame variable name
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.title('Box Plot of Math Scores by Gender')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='writing score', y='math score', data=pdsdf, hue='gender', palette='mako', alpha=0.9) # Changed df to pdsdf to match the DataFrame variable name
plt.xlabel('writing score ')
plt.ylabel('math Score')
plt.title('Scatter Plot of Writing vs. Math Scores')
plt.grid(True)
plt.legend(title='Gender')
plt.show()
```

```python
a = pdsdf['parental level of education'].value_counts() # Changed df to pdsdf to match the DataFrame variable name
plt.figure(figsize=(8, 6))
plt.pie(a, labels=a.index, autopct='%1.1f%%', colors=['red', 'black', 'orange', 'pink', 'green', 'violet'])
plt.title('Pie Chart of Parental level of Education Types')
plt.show()
```
